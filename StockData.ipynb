```python
import yfinance as yf
import pandas as pd
import os

ticker = 'TITAN.NS'

# 1. Download the data
df = yf.download(ticker, start='2023-07-05', end='2025-07-04')

# 2. Select required columns
df = df[['Open', 'High', 'Low', 'Close']].reset_index(drop=True)

# 3. Proceed with transformation
window_size = 44
output_csv = 'stock_training_data.csv'
data_rows = []

for i in range(len(df) - window_size - 1):
    row = {}
    row['Ticker'] = ticker.replace('.NS', '')
    for j in range(window_size):
        row[f'day{j+1}_open'] = float(df.loc[i+j, 'Open'])
        row[f'day{j+1}_high'] = float(df.loc[i+j, 'High'])
        row[f'day{j+1}_low'] = float(df.loc[i+j, 'Low'])
        row[f'day{j+1}_close'] = float(df.loc[i+j, 'Close'])
    row['today_open'] = float(df.loc[i+window_size, 'Open'])
    row['today_high'] = float(df.loc[i+window_size, 'High'])
    row['today_low'] = float(df.loc[i+window_size, 'Low'])
    row['today_close'] = float(df.loc[i+window_size, 'Close'])
    
    data_rows.append(row)

train_df = pd.DataFrame(data_rows)
```

```python
train_df.head()
```

```python
train_df.to_csv(output_csv, mode='a', index=False, header=not os.path.exists(output_csv))
```
